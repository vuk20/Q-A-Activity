### 4.Read the given data set and answer the following questions. Show the steps taken to produce your final answer.

1. Display the Glucose level and the Age for the samples having the index ranging from 45 to 60.
2. How many patients in the data are having the blood pressure level above 120?
3. Select the rows the age is between 24 and 29 (inclusive)
4.Find the minimum and maximum value of insulin level in the data.
5. Create a histogram of "Age".
6. Create a scatter plot of "BMI" vs "Glucose".
7. Replace the min value of "BMI" with mean value of BMI
8. Draw a grouped boxplot with "Outcome" on x-axis and "Glucose" on the y-axis such that we see distributions of Glucose separately for each outcome.

```python
import pandas as pd
data = pd.read_csv('diabetes_model.csv')
data
```

```python
glucose_age=data[45:61]
glucose_age[['Glucose','Age']]
```

```python
count=0
for i in data.BloodPressure:
    if (i > 120):
        count += 1
print('Number of patients having the blood pressure level above 120 in the given data =',count)
```

```python
data[(data.Age>=24) & (data.Age<=29)]
```

```python
print('Minimum value of insulin level in the data =',data.Insulin.min())
print('Maximun value of insulin level in the data =',data.Insulin.max())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
plt.hist(data['Age'],rwidth=0.8)
plt.title('Age distribution of Diabetic patients')
plt.xlabel('Age')
plt.ylabel('Frequency')
```

```python
plt.figure(figsize=(10,10))
plt.scatter(data['BMI'],data['Glucose'],s=12,c='Red')
plt.title('BMI vs Glucose')
plt.xlabel('BMI')
plt.ylabel('Glucose')
```

```python
data = pd.read_csv('diabetes_model.csv')
min_BMI = data.BMI.min()
mean_BMI = round(data.BMI.mean(),2)
new_data = pd.DataFrame(data)
new_data['BMI'] = data['BMI'].replace(min_BMI,mean_BMI) 
new_data[new_data.BMI==mean_BMI]
```

```python
import seaborn as sns

sns.boxplot(x='Outcome',y='Glucose',data=data)
plt.title('Distribution of Glucose for each Outcome',fontsize="14")
```

### 5.

**-** Create a numpy array (first) with 5 x 5 size and all elements are 0 in that.

**-** Create a numpy array (second) with 5 x 5 and all elements are 1 in that.

**-** Combine the arrays first and second vertically, the final array size should be 10 x 5

```python
first_arr = np.zeros((5,5))
print('First array with 5 x 5 size and all elements as 0\n', first_arr)
second_arr = np.ones((5,5))
print('\nSecond array with 5 x 5 size and all elements as 1\n', second_arr)
final_arr = np.concatenate((first_arr,second_arr))
print('\nFinal combined array of First and Second array vertically is\n', final_arr)
```
